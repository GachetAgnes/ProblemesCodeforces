```python
a = input()
b = int(input())
data = []
for i in range(b):
    c = input()
    c = c.split()
    data.append(c)
```

```python
def result(data, main):
    result = []
    for i in data:
        indexes = []
        v = list(i[0])
        for j in v:
            indexes.append(main.find(j))
            print(f'{j} {indexes} {max(indexes)}')
        result.append(max(indexes)+1)
    return result
```

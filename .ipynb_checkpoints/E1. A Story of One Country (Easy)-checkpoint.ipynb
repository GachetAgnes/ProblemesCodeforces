```python
no_chateaux = 4
coord_chateaux = {}

# 0 0 1 2
# 0 2 1 3
# 1 0 2 1
# 1 1 2 3
```

```python
for i in range(no_chateaux):
    a = input()
    a = a.split()
    a = list(map(int, a))
    coord_chateaux[f'Chateau {i}'] = a
```

```python
for key in coord_chateaux:
    print(f'{key} a la valeur : {coord_chateaux[key]}')
```

```python
for key in coord_chateaux:
    a = [] 
    b = []
    a = coord_chateaux[key[:2]]
    b = coord_chateaux[key[2:4]]
    c = (a, b)
    print(f'{key} a la valeur : {c}')
```

```python
# Variables
temps_total = 15
no_eleves = 7
temps_eleves = [1, 2, 3, 4, 5, 6, 7]
 

temps_total2 = 5
no_eleves2 = 100
temps_eleves2 = [80, 40, 40, 40, 60]
```

```python
def eleves(temps_total, no_eleves, temps_eleves):
    total = 0
    index = 0
    
    for i in temps_eleves:
        total += i
        if total > temps_total:
            break
        index += 1
    return temps_eleves[index : no_eleves]
    
```

```python
a = []
a = eleves(temps_total, no_eleves, temps_eleves)
print(a)
```

```python
def composants(a, temps_eleves):
    resultat = {}
    for i in a:
        resultat_partiel = []
        index = 0
        total = 0
        temps_restant = temps_eleves[index:]
        for j in temps_restant:
            total += j
            if total >= i:
                resultat_partiel.append(index)
            index += 1
        
        # print(resultat_partiel)
        if resultat_partiel:
            resultat[i] = min(resultat_partiel)  
        
    return resultat
```

```python
b = composants(a, temps_eleves)
print(b)
```

```python
import random as rd

def test_data():
    temps_total = rd.randint(0, 100)
    no_eleves = rd.randint(0, 100)
    temps_eleves = []
    for i in range(0, no_eleves):
        temps_eleves.append(rd.randint(1, temps_total/2))
        
    return (temps_total, no_eleves, temps_eleves)
```

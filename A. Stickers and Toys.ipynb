```python
a = int(input())
data = []
for i in range(a):
    b = input()
    b = b.split()
    b = list(map(int, b))
    data.append(b)
print(data)
```

```python
def result(data):
    result = []
    for i in data:
        max_eggs = i[0]
        s = i[1]
        t = i[-1]
        if max_eggs == s and s == t or max_eggs == s or max_eggs == t:
            result.append(1)
        else:
            result.append(min(s, t)+1)
    return result
```

```python
print(result(data))
```
